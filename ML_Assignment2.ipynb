# Part **2**

```python
#2
import pandas as pd
df=pd.read_csv('AWCustomers.csv')
print(pd.isna(df))
```

```python
df.drop(['Title'],axis=1,inplace=True)
df.drop(['Suffix'],axis=1,inplace=True)
df.drop(['AddressLine2'],axis=1,inplace=True)
df
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
df['MiddleName'] = df['MiddleName'].fillna(df['MiddleName'].mode().iloc[0])
df
```

# **Part 2 (b)**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df['YearlyIncome']=scaler.fit_transform(df['YearlyIncome'].values.reshape(-1,1)).flatten()
df
```

# **Part 2 (c)**

```python
import pandas as pd
import numpy as np
bins=np.linspace(min(df['YearlyIncome']),max(df['YearlyIncome']),4)
group_names=['Level1','Level2','Level3']
df['YearlyIncome']=pd.cut(df['YearlyIncome'],bins,group_names,include_lowest=True)
df
```

# **Part 2 (d)**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['NumberCarsOwned'] = scaler.fit_transform(df['NumberCarsOwned'].values.reshape(-1, 1)).flatten()
df
```

# **Part 2 (e)**

```python
from sklearn.preprocessing import OneHotEncoder
columns = ['Education',	'Occupation',	'Gender',	'MaritalStatus'	]

encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(df[columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(columns))
df_encoded = pd.concat([df, one_hot_df], axis=1)
df_encoded = df_encoded.drop(columns, axis=1)
df_encoded
```

# **Part 3**

```python
a =  df['TotalChildren']
b = df['HomeOwnerFlag']
c = list(set(a) & set(b))
d = list(set(a) | set(b))
J = float(len(c)+len(d)/len(d))
print("Simple Matching :",J)
```

```python
a= df['TotalChildren']
b= df['HomeOwnerFlag']
c= list(set(a) & set(b))
d= list(set(a) | set(b))
J=float(len(c)/len(d))
print("Jaccard Similarity :",J)
```

```python
import numpy as np
from numpy.linalg import norm
a =  df['TotalChildren']
b = df['HomeOwnerFlag']
cosine = np.dot(a,b)/(norm(a)*norm(b))
print("Cosine Similarity :",cosine)
```
